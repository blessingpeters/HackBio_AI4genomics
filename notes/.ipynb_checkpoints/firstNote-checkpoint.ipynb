Grouping and summarising data


```python
expr_long = pd.DataFrame({
    "Gene": ["GeneA","GeneA", "GeneB", "GeneB", "GeneC", "GeneC" ],
    "Conditions": ["Control","Treatment", "Control","Treatment", "Control","Treatment", ],
    "Expression": [5.2, 6.1, 2.1, 3.3, 7.8, 5.9]
})
expr_long
```

average expression per gene

```python
expr_long.groupby("Gene")["Expression"].mean()
```

average expression per condition

```python
expr_long.groupby("Conditions")["Expression"].mean()
```

more summary at once

```python
expr_long.groupby("Conditions")["Expression"].agg(["mean", "min", "max", "std"])
```
